```python
import cv2
```

```python
with open("removed.txt", "w") as file:
    for i in range(0, 6):
        for j in range(1, 180):
            filename = 'men/{}/{}_men ({}).JPG'.format(i, i , j)
            print(filename)
            img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
            if img is None:
                file.write(filename + "\n")
            
            filename = 'Woman/{}/{}_woman ({}).JPG'.format(i, i, j)
            img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
            if img is None:
                file.write(filename + "\n")
            print(filename)    
               
```
